```python
#I used this code to check for missing values in the DataFrame df using the isnull() method, which returns a DataFrame of the same shape as df, but with Boolean values indicating whether each element is missing or not.
#The output of df.isnull().sum() shows the number of missing values in each column of the DataFrame df.
#it shows that the columns education, cigsPerDay, BPMeds, totChol, BMI, heartRate, and glucose have missing values. The numbers in the output indicate the number of missing values in each column.
#It's important to handle missing values appropriately before performing any analysis or modeling on the data, as missing values can impact the accuracy and validity of the results.
```

```python
df['education'].fillna(value=df['education'].mode()[0], inplace=True)
df['cigsPerDay'].fillna(value=df['cigsPerDay'].mode()[0], inplace=True)
df['BPMeds'].fillna(value=df['BPMeds'].mode()[0], inplace=True)
df['totChol'].fillna(value=df['totChol'].mode()[0], inplace=True)
df['BMI'].fillna(value=df['BMI'].mode()[0], inplace=True)
df['heartRate'].fillna(value=df['heartRate'].mode()[0], inplace=True)
df['glucose'].fillna(value=df['glucose'].mode()[0], inplace=True)
```
